```python
#tokenizer
from nltk.tokenize import word_tokenize,sent_tokenize
sents = nltk.sent_tokenize(paragraph)
sents
```

```python
words= nltk.word_tokenize(paragraph)
words
```

```python
unique_tokens = set(words)
unique_tokens
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
final_tokens = []
for word in words:
    if word not in stop_words:
        final_tokens.append(word)
print(f"The final number of tokens after removing stopwords are {len(final_tokens)}")
```

```python
#stemming means removing of prefix and suffix of similar words
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer

#snowball stemmer has a language parameter
words = ["grows","leaves","fairly","cats","trouble","misunderstanding","friendships","easily", "rational", "relational"]

#creating instances of both stemmer
ps = PorterStemmer()
stemmed_words_ps = [ps.stem(word) for word in words]
print("PorterStemmed words",stemmed_words_ps)
```

```python
#snowball stemmer
snow_s = SnowballStemmer("english")
stemmed_words_snow_s= [snow_s.stem(word) for word in words]
print("Snowball stemmer", stemmed_words_snow_s)
```

## Lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
#since lemmatization is based on wordnets built in morph function
```

```python
words = ["grows","leaves","fairly","cats","trouble","running","friendships","easily", "was", "relational","has"]
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
print("Lemmatization of words without using POS tag")
for i,j in zip(words,lemmatized_words):
    print(i,"-",j)
```

```python
#lemmatized word using part of speech
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word,pos = "v") for word in words]
print("Lemmatization of words using POS tag")
for i,j in zip(words,lemmatized_words):
    print(i,"-",j)
```
